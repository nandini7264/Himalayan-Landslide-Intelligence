- This file is for scraping landslide data in Indian and Nepali Himalayan regions.
- Extracted data from the NASA COOLR website that has open landslide data from around the world.
- Used ArcGIS API service to query and extract useful landslide data from COOLR database
- Queried data for only Indian and Nepali Himalayan regions
- Converted data into dataframe and saved it in a csv file(landslides_data.csv)

```python
# importing libraries
import requests
import pandas as pd
import numpy as np
```

### Data Extraction using API Pagination

```python
# COOLR website url to extract data
url = "https://gis.earthdata.nasa.gov/gis05/rest/services/Landslides/COOLR_Reports_Points/FeatureServer/0/query"

all_records = []    # empty list to store extracted data

offset = 0    # pagination to extract multiple rows of data
batch_size = 2000

while True:

    params = {    # creating instructions for API
        "where": "country_name IN ('India', 'Nepal')",    # querying data from India and Nepal only
        "outFields": "*",    # all columns
        "f": "json",    # json format
        "resultOffset": offset,
        "resultRecordCount": batch_size    # send 2000 records only
    }

    response = requests.get(url, params=params)
    data = response.json()

    features = data.get("features", [])

    if not features:
        break

    for feature in features:
        all_records.append(feature["attributes"])    # extract attributes of landslides for each landslide

    print(f"Fetched {len(features)} records")

    offset += batch_size
```

```python
# converting list of landslides into a dataframe
df = pd.DataFrame(all_records)
df.head()
```
